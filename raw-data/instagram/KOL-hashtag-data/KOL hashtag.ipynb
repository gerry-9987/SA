# Do a sanity check on her followers

- Make sure she has real followers?

```python
accounts = [1,2,3,4,5,6]
account = accounts[0]
USER = config(f'USER{account}',default='')
PASSWORD = config(f'PASSWORD{account}',default='')
print(USER)
L = instaloader.Instaloader()
L.login(USER, PASSWORD)
```

```python
def create_data():
    data = {}
    for i in ['username','mediacount','followers','followees','biography','profile_pic_url']:
        data[i]=[]
    return data
```

```python
username = 'ss_rockmyworld'
file_name = f"{username}.csv"

profile = instaloader.Profile.from_username(L.context, username)
followers = profile.get_followers()
follower_count = profile.followers

# create dataset
if exists(file_name):data = pd.read_csv(file_name).iloc[:,1:].to_dict('list')
else:create_data()

# Loop the followers 
for idx, follower in enumerate(followers):
    
    username = follower.username
    print(f'doing {username} now')
    if username in data['username']:continue
    data['username'].append(follower.username)
    data['mediacount'].append(follower.mediacount)
    data['followers'].append(follower.followers)
    data['followees'].append(follower.followees)
    data['biography'].append(follower.biography)
    data['profile_pic_url'].append(follower.profile_pic_url)
    pd.DataFrame(data=data).to_csv(file_name)
    clear_output(wait=True)
    print(f"{idx}/{follower_count} done for {username}")
```
