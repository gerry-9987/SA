```python
import pandas as pd
```

```python
df_covet = pd.read_csv("raw-data/instagram/social-netork/competitors_details/2.0/covetcrystalsjewelry.csv")
df_crystlmojo = pd.read_csv("raw-data/instagram/social-netork/competitors_details/2.0/crystalmojo.csv")
df_fateditis = pd.read_csv("raw-data/instagram/social-netork/competitors_details/2.0/fated.itis.csv")
df_gemstory = pd.read_csv("raw-data/instagram/social-netork/competitors_details/2.0/gemstorysg.csv")
df_glacecrystals = pd.read_csv("raw-data/instagram/social-netork/competitors_details/2.0/glacecrystals.csv")
df_illanocte = pd.read_csv("raw-data/instagram/social-netork/competitors_details/2.0/illa_nocte.csv")
df_kaylacrystals = pd.read_csv("raw-data/instagram/social-netork/competitors_details/2.0/kayla.crystals.csv")
df_loveforcrystals = pd.read_csv("raw-data/instagram/social-netork/competitors_details/2.0/loveforcrystals.csv")
df_ohmou = pd.read_csv("raw-data/instagram/social-netork/competitors_details/2.0/ohmou.sg.csv")
df_soulalign = pd.read_csv("raw-data/instagram/social-netork/competitors_details/2.0/soulalign.csv")
df_tamza = pd.read_csv("raw-data/instagram/social-netork/competitors_details/2.0/tamzacrystals.csv")
df_jamstones = pd.read_csv("raw-data/instagram/js-posts-all/optimal_post_timing2.0.csv")
```
