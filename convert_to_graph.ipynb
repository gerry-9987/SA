```python


import networkx as nx
from networkx.algorithms import community
from networkx.readwrite import json_graph
import json

from IPython.display import Image
import matplotlib.pyplot as plt
import itertools
```

```python
def read_json_file(filename):
    with open(filename) as f:
        js_graph = json.load(f)
    return json_graph.node_link_graph(js_graph)
```

```python
starwars = read_json_file("../star-wars-network-data-master/starwars-full-interactions.json")
```

```python
f = open('../star-wars-network-data-master/starwars-full-interactions.json')

data = json.load(f)

labeldict = {}

for i in range(len(data['nodes'])):
    # print(i)
    labeldict[i] = data['nodes'][i]["name"]
 
# # Closing file
f.close()

print(labeldict)
```

```python
pos = nx.spring_layout(starwars)
nx.draw(starwars, cmap = plt.get_cmap('rainbow'), with_labels=True, pos=pos)
```

```python
nx.draw(starwars, cmap = plt.get_cmap('rainbow'), labels=labeldict, pos=pos)
```

```python
nx.write_gml(starwars, "starwars2.gml")
```
