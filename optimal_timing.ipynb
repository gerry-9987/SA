Optimal Timing


```python
import pandas as pd
import numpy as np
import datetime
from dateutil import parser
```

```python
df=pd.read_csv("Crystal Hashtags Data/instagram_crystalsg_hashtag_clean_new.csv")
```

```python
df.head()
```

```python
df1=df[["username","fullName","pubDate","commentCount","likeCount", "description","clean_description","clean_tokens"]]
df1.head(10)
```

```python
from datetime import date
```

```python
df1["Date"] = df1["pubDate"].apply(lambda x: parser.parse(x))
df1["Day of Week"] = df1["Date"].apply(lambda x: x.weekday())
df1["Day of Week"] = df1["Date"].dt.day_name()
df1["Hour of Day"] = df1["Date"].apply(lambda x: x.hour)
```

```python
df1.sort_values(['likeCount'], ascending=False).head(20)
```

```python
df_all_posts_usernames = df1.sort_values(['likeCount'], ascending=False)

df_all_posts_usernames
```

```python
df_all_posts_mean = df1.groupby(["Day of Week", "Hour of Day"]).mean().sort_values(by="likeCount",ascending=False)
df_all_posts_mean 
# df2 = df1.groupby(["Day of Week", "Hour of Day"]).count().sort_values(by="likeCount",ascending=False)
```

```python
df_all_posts_mean_sort_commentCount = df1.groupby(["Day of Week", "Hour of Day"]).mean().sort_values(by="commentCount",ascending=False)
df_all_posts_mean_sort_commentCount 
```

```python
df_all_posts_sum  = df1.groupby(["Day of Week", "Hour of Day"]).sum().sort_values(by="likeCount",ascending=False)
df_all_posts_sum
```

```python
df_all_posts_count = df1.groupby(["Day of Week", "Hour of Day"]).count().sort_values(by="likeCount",ascending=False)

df_all_posts_count
```
