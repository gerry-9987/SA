```python
import pandas as pd
import numpy as np
import datetime
from dateutil import parser
import matplotlib as plt
from datetime import date
```

```python
df=pd.read_csv("Crystal Hashtags Data/instagram_crystalsg_hashtag_clean_new.csv")
```

```python
df.head(10)
```

```python
df1=df[["username","fullName","pubDate","commentCount","likeCount", "description","clean_description","clean_tokens"]]
df1.head(10)
```

```python
df1["Date"] = df1["pubDate"].apply(lambda x: parser.parse(x))
df1["Day of Week"] = df1["Date"].apply(lambda x: x.weekday())
df1["Day of Week"] = df1["Date"].dt.day_name()
df1["Hour of Day"] = df1["Date"].apply(lambda x: x.hour)
```

```python
df1_no_giveaway = df1[df1["clean_tokens"].str.contains("giveaway") == False]

df1_no_giveaway
```

```python
df1_giveaway = df1[df1["clean_tokens"].str.contains("giveaway") == True]

df1_giveaway
```

```python
# not very useful cuz sale can mean promotion or to sell crystals


df1_sale = df1[df1["clean_tokens"].str.contains("sale") == True]

df1_sale
```

```python
df1_giveaway = df1_giveaway.sort_values(['likeCount'], ascending=False)

df1_giveaway
```

```python
df1_no_giveaway = df1_no_giveaway.sort_values(['likeCount'], ascending=False)

df1_no_giveaway
```
