```python
import pandas as pd
import numpy as np
import datetime
from dateutil import parser
import matplotlib as plt
from datetime import date
```

```python
df=pd.read_csv("Crystal Hashtags Data/instagram_crystalsg_hashtag_clean_new.csv")
```

```python
df.head(10)
```

```python
df1=df[["username","fullName","pubDate","commentCount","likeCount", "description","clean_description","clean_tokens"]]
df1.head(10)
```

```python
df1["Date"] = df1["pubDate"].apply(lambda x: parser.parse(x))
df1["Day of Week"] = df1["Date"].apply(lambda x: x.weekday())
df1["Day of Week"] = df1["Date"].dt.day_name()
df1["Hour of Day"] = df1["Date"].apply(lambda x: x.hour)
```

```python
df1_no_giveaway = df1[df1["clean_tokens"].str.contains("giveaway") == False]

df1_no_giveaway
```

```python
df1_giveaway = df1[df1["clean_tokens"].str.contains("giveaway") == True]

df1_giveaway
```

```python
# not very useful cuz sale can mean promotion or to sell crystals


df1_sale = df1[df1["clean_tokens"].str.contains("sale") == True]

df1_sale
```

```python
df1_no_giveaway.drop(index=df1_no_giveaway.index[0], 
        axis=0, 
        inplace=True)

df1_no_giveaway
```

```python
df1_no_giveaway_mean_sort_likes = df1_no_giveaway.groupby(["Day of Week", "Hour of Day"]).mean().sort_values(by="likeCount",ascending=False)
```

```python
df1_no_giveaway_mean_sort_comments = df1_no_giveaway.groupby(["Day of Week", "Hour of Day"]).mean().sort_values(by="commentCount",ascending=False)
```

```python
df1_giveaway_mean_sort_likes = df1_giveaway.groupby(["Day of Week", "Hour of Day"]).mean().sort_values(by="likeCount",ascending=False)
```

```python
df1_giveaway_mean_sort_comments = df1_giveaway.groupby(["Day of Week", "Hour of Day"]).mean().sort_values(by="commentCount",ascending=False)
```

```python
ax_df1_no_giveaway_mean_sort_likes = df1_no_giveaway_mean_sort_likes.head(20).plot.bar(y=["commentCount","likeCount"], figsize=(20,18))

for container in ax_df1_no_giveaway_mean_sort_likes.containers:
    ax_df1_no_giveaway_mean_sort_likes.bar_label(container)
```

```python
ax_df1_giveaway_mean_sort_likes = df1_giveaway_mean_sort_likes.head(20).plot.bar(y=["commentCount","likeCount"], figsize=(20,18))

for container in ax_df1_giveaway_mean_sort_likes.containers:
    ax_df1_giveaway_mean_sort_likes.bar_label(container)
```

```python
ax_df1_no_giveaway_mean_sort_comments = df1_no_giveaway_mean_sort_comments.head(20).plot.bar(y=["commentCount","likeCount"], figsize=(20,18))

for container in ax_df1_no_giveaway_mean_sort_comments.containers:
    ax_df1_no_giveaway_mean_sort_comments.bar_label(container)
```
